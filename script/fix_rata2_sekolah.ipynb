```python
import mysql.connector
import pandas as pd 
import json
import numpy as np
import MySQLdb
import MySQLdb.cursors
```

```python
db_thread = MySQLdb.connect("127.0.0.1", "root", "andhika", "usek_monev", cursorclass=MySQLdb.cursors.DictCursor)
cur_thread = db_thread.cursor()
print(cur_thread)
```

```python
def get_sekolah(kurikulum):
    sql = "SELECT * FROM sekolah where kurikulum='%s'" % (kurikulum)
    cur_thread.execute(sql)
    sekolah = cur_thread.fetchall()
    return sekolah
```

```python
def get_user_sekolah(sekolah_id):
    sql = "SELECT * FROM user where sekolah_id='%s' and level = 'siswa'" % (sekolah_id)
    cur_thread.execute(sql)
    user = cur_thread.fetchall()
    return user
```

```python
def get_ujian_siswa(siswa_id):
    sql = "select * from ujian_siswa where user_id='%s' and status > 1" % (siswa_id)
    ujian_siswa = pd.read_sql(sql, con=db_thread)
    ujian_siswa = ujian_siswa.set_index('id') 
    return ujian_siswa
```
