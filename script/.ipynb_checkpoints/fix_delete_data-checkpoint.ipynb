```python
sql = "SELECT count(*) as jumlah, u.sekolah_id from `user` u join sekolah s on u.sekolah_id = s.id WHERE s.kurikulum = '2013' and u.jurusan = 'MIPA' and u.`level` = 'siswa' group by u.sekolah_id order by jumlah desc;"
cur_thread.execute(sql)
sekolahs = cur_thread.fetchall()
```

```python
for sekolah in sekolahs:
    if sekolah['jumlah'] < 11:
        print('sekolah_id: ', sekolah['sekolah_id'])        
        print('jumlah: ', sekolah['jumlah'])
        jumlah = int(sekolah['jumlah']-0)
        sql = "select * from user u join sekolah s on u.sekolah_id = s.id where s.kurikulum = '2013' and u.level='siswa' and u.sekolah_id='%s' and u.jurusan = 'MIPA' order by u.id" %(sekolah['sekolah_id'])
        cur_thread.execute(sql)
        siswas = cur_thread.fetchall()
        siswa_id = []
        for siswa in siswas:
            siswa_id.append(siswa['id'])
        siswa_id_tuple = tuple(siswa_id)
        sql = "delete from ujian_siswa where user_id in {};".format(siswa_id_tuple)
        cur_thread.execute(sql)
        db_thread.commit()
        sql = "delete from user where level='siswa' and id in {};".format(siswa_id_tuple)
        cur_thread.execute(sql)
        db_thread.commit()
        print("ok")
```
